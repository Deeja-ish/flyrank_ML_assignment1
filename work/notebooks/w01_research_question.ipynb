## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd 

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")


total_pages = df.shape[0]
total_clients = df['client_id'].unique()
declining_pages = (df['trend_direction'] == 'down').sum()
percentage = (declining_pages / total_pages) * 100

print(f"Total pages (rows): {total_pages:,}")
print(f"Distinct client IDs: {total_clients}")
print(f" Pages in decline: {declining_pages:,} ({percentage:.1f}%)")
```
