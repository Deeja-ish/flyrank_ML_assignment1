## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A manual rule (e.g., "if age > 180 days and CTR < 1%") is too simple. In reality, page health depends on many interacting signals: search volume, keyword competition, user engagement (scroll rate, sessions), LLM model context, and intent. 
A human editor cannot write nested if-statements to weigh 44 different signals and find non-linear relationships across 32 clients. 
A machine learning model (such as a Random Forest) does this automatically and adapts to patterns in the data.
